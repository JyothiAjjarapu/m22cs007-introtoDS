```python
import pandas as pd
from tabulate import tabulate


jyothi_df3 = pd.read_csv(file_path)

# Clean the dataset: Convert columns to numeric and handle missing values
numeric_columns = ['TotalSamples', 'Negative', 'Positive']
jyothi_df3[numeric_columns] = jyothi_df3[numeric_columns].apply(pd.to_numeric, errors='coerce')

# Calculate central tendency measures for numeric columns
central_tendency_measures = []
for column in numeric_columns:
    mean = jyothi_df3[column].mean()
    median = jyothi_df3[column].median()
    mode = jyothi_df3[column].mode().iloc[0] if len(jyothi_df3[column].mode()) > 0 else None
    central_tendency_measures.append([column, mean, median, mode])

# Tabulate the central tendency measures
table_headers = ['Column', 'Mean', 'Median', 'Mode']
print(tabulate(central_tendency_measures, headers=table_headers, tablefmt='pretty'))
```

```python
measures = []
for column in numeric_columns:
    column_data = jyothi_df3[column].dropna()
    data_range = column_data.max() - column_data.min()
    data_variance = column_data.var()
    data_std_dev = column_data.std()
    measures.append([column, data_range, data_variance, data_std_dev])

table_headers = ['Column', 'Range', 'Variance', 'Standard Deviation']
print(tabulate(measures, headers=table_headers, tablefmt='pretty'))
```

```python
coeff_of_var_measures = []
for column in numeric_columns:
    column_data = jyothi_df3[column].dropna()
    data_std_dev = column_data.std()
    coeff_of_var = (data_std_dev / column_data.mean()) * 100
    coeff_of_var_measures.append([column, coeff_of_var])

table_headers = ['Column', 'Coefficient of Variation (%)']
print(tabulate(coeff_of_var_measures, headers=table_headers, tablefmt='pretty'))
```

```python
import pandas as pd
import numpy as np

# Load your dataset into a DataFrame named jyothi_df3

# List of numeric columns in the dataset
numeric_columns = ['TotalSamples', 'Negative', 'Positive']

# Initialize a list to store IQR and quartiles measures
iqr_quartiles_measures = []

# Calculate IQR and quartiles for each numeric column
for column in numeric_columns:
    column_data = jyothi_df3[column].dropna()
    q1 = column_data.quantile(0.25)
    q3 = column_data.quantile(0.75)
    data_iqr = q3 - q1
    iqr_quartiles_measures.append([column, data_iqr, q1, q3])

# Tabulate IQR and quartiles measures
table_headers = ['Column', 'IQR', 'Q1', 'Q3']
print(tabulate(iqr_quartiles_measures, headers=table_headers, tablefmt='pretty'))
```

```python
import pandas as pd
import numpy as np
from scipy.stats import skew, kurtosis

# Load your dataset into a DataFrame named jyothi_df3

# List of numeric columns in the dataset
numeric_columns = ['TotalSamples', 'Negative', 'Positive']

# Initialize a list to store skewness and kurtosis measures
skew_kurtosis_measures = []

# Calculate skewness and kurtosis for each numeric column
for column in numeric_columns:
    column_data = jyothi_df3[column].dropna()
    data_skewness = skew(column_data)
    data_kurtosis = kurtosis(column_data)
    skew_kurtosis_measures.append([column, data_skewness, data_kurtosis])

# Tabulate skewness and kurtosis measures
table_headers = ['Column', 'Skewness', 'Kurtosis']
print(tabulate(skew_kurtosis_measures, headers=table_headers, tablefmt='pretty'))

```
