```python
# List of numerical columns for which to calculate central tendency
numerical_columns = ['Confirmed', 'Cured', 'Deaths']  # Add more columns if needed

for column_name in numerical_columns:
    mean = data[column_name].mean()
    median = data[column_name].median()
    mode = data[column_name].mode().iloc[0]
    
    print(f"Central Tendency Measures for '{column_name}':")
    print("Mean:", mean)
    print("Median:", median)
    print("Mode:", mode)
    print()
```

```python

# List of numerical columns for which to calculate dispersion and consistency
numerical_columns = ['Confirmed', 'Cured', 'Deaths']  # Add more columns if needed

dispersion_consistency_data = []

for column_name in numerical_columns:
    column_data = data[column_name]
    
    range_value = column_data.max() - column_data.min()
    variance = column_data.var()
    std_deviation = column_data.std()
    coefficient_of_variation = (std_deviation / column_data.mean()) * 100
    
    dispersion_consistency_data.append({
        'Column': column_name,
        'Range': range_value,
        'Variance': variance,
        'Standard Deviation': std_deviation,
        'Coefficient of Variation': coefficient_of_variation
    })

# Create a DataFrame from the collected data
dispersion_consistency_df = pd.DataFrame(dispersion_consistency_data)

# Display the tabulated summary
print(dispersion_consistency_df)
```

```python

# List of numerical columns for which to calculate IQR
numerical_columns = ['Confirmed', 'Cured', 'Deaths']  # Add more columns if needed

iqr_data = []

for column_name in numerical_columns:
    column_data = data[column_name]
    
    q1 = column_data.quantile(0.25)
    q3 = column_data.quantile(0.75)
    iqr = q3 - q1
    
    iqr_data.append({
        'Column': column_name,
        'IQR': iqr
    })

# Create a DataFrame from the collected data
iqr_df = pd.DataFrame(iqr_data)

# Display the tabulated summary
print(iqr_df)
```

```python
from scipy.stats import skew

# Calculate skewness for each column
skewness_data = []

for column_name in numerical_columns:
    column_data = data[column_name]
    skew_value = skew(column_data)
    
    skewness_data.append({
        'Column': column_name,
        'Skewness': skew_value
    })

# Create a DataFrame from the collected data
skewness_df = pd.DataFrame(skewness_data)

# Display the tabulated summary
print(skewness_df)
```

```python
from scipy.stats import kurtosis

# Calculate kurtosis for each column
kurtosis_data = []

for column_name in numerical_columns:
    column_data = data[column_name]
    kurtosis_value = kurtosis(column_data)
    
    kurtosis_data.append({
        'Column': column_name,
        'Kurtosis': kurtosis_value
    })

# Create a DataFrame from the collected data
kurtosis_df = pd.DataFrame(kurtosis_data)

# Display the tabulated summary
print(kurtosis_df)
```

```python
jyothi_df1.drop(['Time','Sno','ConfirmedIndianNational','ConfirmedForeignNational'],inplace =True, axis = 1)
```

```python
jyothi_df1['ActiveCases'] = jyothi_df1['Confirmed'] - (jyothi_df1['Cured'] + jyothi_df1['Deaths'])
jyothi_df1.tail()
```

```python
top_10_active_cases = jyothi_df1.groupby(by = 'State/UnionTerritory').max()[['ActiveCases' , 'Date']].sort_values(by='ActiveCases',ascending = False).reset_index()
fig = plt.figure(figsize=(18,8))
ax = sns.barplot(data = top_10_active_cases.iloc[:10], y='ActiveCases', x='State/UnionTerritory')
plt.title('Top 10 states with most active cases',size = 25)
plt.xlabel('State')
plt.ylabel('Active Cases')
```

```python
top_10_death_cases = jyothi_df1.groupby(by = 'State/UnionTerritory').max()[['Deaths' , 'Date']].sort_values(by='Deaths',ascending = False).reset_index()
fig = plt.figure(figsize=(18,8))
ax = sns.barplot(data = top_10_death_cases.iloc[:12], y='Deaths', x='State/UnionTerritory')
plt.title('Top 10 states with most death cases',size = 25)
plt.xlabel('State')
plt.ylabel('Deaths')
```

```python
fig = plt.figure(figsize=(12,6))
ax = sns.lineplot(data = jyothi_df1[jyothi_df1['State/UnionTerritory'].isin(['Maharashtra','Karnataka','Kerala','Tamil Nadu','Uttar Pradesh'])],x='Date',y='ActiveCases',hue = 'State/UnionTerritory')

ax.set_title("Top 5 Affected States in India",size = 16)
```
